```python
def age_group(N):
    a = min(data['Age'])
    b = max(data['Age'])
    p = (b-a)/(N-1)
    group =  data['Age'].apply(lambda x: next((str(round(a + (k-1)*p)) + '-' + str(round(a + k*p)) for k in range(N) if x in range(round(a + (k-1)*p), round(a + k*p)+1)), 'out'))    
    return group
```

```python
data['group']=age_group(4)
```

```python
sub = make_subplots(rows=1, cols=5, specs=[[{'type': 'domain'}, {'type': 'domain'}, {'type': 'domain'}, {'type': 'domain'}, {'type': 'domain'}]])

# Titles for each pie chart
# titles = ['Age 18-24', 'Age 25-34', 'Age 35-44', 'Age 45-54']

# Create Plotly Express figures for each age band
for i, age_band in enumerate(list_age):
    p = px.pie(data[data['CF_age band'] == age_band], values='depart', names='Gender', labels={'depart': 'Depart'}, hole=0.5)
    for trace in p.data:
        sub.add_trace(trace, row=1, col=i + 1)

    # Calculate total for the current age band
    total = data[data['CF_age band'] == age_band]['depart'].sum()

    # Calculate x-coordinate for annotation based on subplot width and position
    x_coord = (2.05* i + 1) / 10-0.01  # Adjusted for equal width of 4 subplots

    # Add annotation for total
    sub.add_annotation(dict(text=f"{total}", x=x_coord, y=0.5, font_size=32, showarrow=False, xref='paper', yref='paper', xanchor='center', yanchor='middle'))

    # Add annotation for title
    sub.add_annotation(dict(text=age_band, x=x_coord, y=-0.04, font_size=15, showarrow=False, xref='paper', yref='paper', xanchor='center', yanchor='middle'))

# Update layout
sub.update_layout(
    title_text='Subplots with Plotly Express Pie Charts',
    height=600,
    width=1200,  # Adjust width as needed
    paper_bgcolor='#1f2c56',
    font_color='white'
)

# Show figure
sub.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd

# Sample data
data = pd.DataFrame({
    'category': ['A', 'B', 'C', 'D', 'E'],
    'value': [4, 7, 1, 8, 5]
})

# Create scatter plot for the lollipops
fig = px.scatter(data, x='category', y='value', size=[10]*len(data), color='value', color_continuous_scale='Viridis')

# Add lines to create the "sticks" of the lollipops
for i, row in data.iterrows():
    fig.add_trace(go.Scatter(
        x=[row['category'], row['category']],
        y=[0, row['value']],
        mode='lines',
        line=dict(color='black', width=2)
    ))

# Update layout
fig.update_layout(
    title='Lollipop Chart',
    xaxis_title='Category',
    yaxis_title='Value',
    showlegend=False
)

# Show figure
fig.show()
```
